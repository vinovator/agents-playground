```python
from crewai import Agent, Task, Crew, Process, LLM
import os
```

```python
local_llm = LLM(
    model="ollama/llama3.1",
    base_url="http://localhost:11434"
)
```

```python
# Define Agents

analyst_features_agent = Agent(
    role = "Features Analyst",
    goal = "Analyse technical specifications",
    backstory = "You focus purely on hardware specs and raw numbers",
    verbose = True,
    llm = local_llm    
)

analyst_pricing_agent = Agent(
    role = "Pricing Analyst",
    goal = "Analyse pricing strategy",
    backstory = "You focus purely on cost, value, and affordability",
    verbose = True,
    llm = local_llm    
)

manager_agent = Agent(
    role = "Manager",
    goal = "Manage the analysis process",
    backstory = "You are a manager who coordinates the work of the analyst agents",
    verbose = True,
    llm = local_llm    
)
```

```python
# Define tasks (Async mode)

task_features = Task(
    description="List the technical specs (CPU, RAM, GPU) of the Macbook Air M3",
    expected_output="A technical data sheet",
    agent=analyst_features_agent,
    async_execution=True # This agent will not block the next one.
)

task_pricing = Task(
    description="List the pricing of the Macbook Air M3",
    expected_output="A pricing table",
    agent=analyst_pricing_agent,
    async_execution=True # This agent will not block the next one.
)

task_synthesis = Task(
    description="Write a summary report combining the specs and pricing",
    expected_output="A final decision memo on whether to buy",
    agent=manager_agent,
    context=[task_features, task_pricing] # Wait for the previous tasks to finish.
    )
# Run tasks
```

```python
# Run Crew
crew = Crew(
    agents=[analyst_features_agent, analyst_pricing_agent, manager_agent],
    tasks=[task_features, task_pricing, task_synthesis],
    verbose=True
)

result = crew.kickoff()
print(result)
```
